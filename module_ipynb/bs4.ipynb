```python
import urllib.request as req
import bs4
```

```python
url = "https://www.ptt.cc/cls/1253"

request = req.Request( url , headers={
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36"
})

with req.urlopen( request ) as indata : 
    data = indata.read().decode( "utf-8" )
    
print(data)
```

```python
soup = bs4.BeautifulSoup( data , "html.parser" )
print( soup.title.string )
```

```python
board = soup.find( "div" , class_="board-title" ) # 找出 class="board-title" 的 div 標籤
print( board )                                    #只有一個
```

```python
board = soup.find_all( "div" , class_="board-title" )
for title in board : 
    print( title.string )
```
