```python
rs = [ 7 , 9 , 6 , 5 , 9 , 9 , 6 , 5 , 5 , 7 , 9 , 8 , 7 ]
animals = ["rat","ox","tiger","rabbit","dragon","snake","horse","goat","monkey","rooster","dog","pig","rat"]

angs = pylab.linspace( 0 , 2*pi , len(animals) )

pylab.polar( angs , rs )

pylab.grid( ls="--" )

for i in range( len(angs) ) : #畫線

    pylab.arrow( 0 , 0 , angs[i] , rs[i] , color = ( random() , random() , random() ) )

pylab.fill( angs , rs , facecolor = "c") #塗滿
pylab.rgrids( arange( 1 , 10 ) , angle = 22 ) #(1到9) 標示在22度
pylab.thetagrids( arange( 0 , 360 , 30 )  ) #標示(0,30,60,90,120,150,180,210,240,270,300,330)度
pylab.xticks( angs ,  animals ) #對應

pylab.show()
```

```python
h = 1e-6

f = lambda x : cos( x )
df = lambda x :  ( f( x + h ) - f( x ) )/h
ddf = lambda x : "x"+str(x)
    
x = linspace( 0 , 1.5*pi , 100 )
y = cos( x )

pylab.title( "Newton method" ) #設定標題
pylab.ylabel( "cos(x)" ) #設定y軸標題
pylab.grid( ls = "--" )

x0 = 0.43

for i in range (1 , 6 ) :

    a = ( random() , random() , random() )

    pylab.arrow( x0 , f( x0 ) , 0 , -f( x0 ) , color = a , ls = "--")
    pylab.text( x0 - 0.1 , -0.1 , ddf( i - 1 ) ) 
    
    x1 = x0 - f( x0 ) / df( x0 )
    pylab.arrow( x0 , f( x0 ) , x1 - x0 , -f( x0 ) , color = a , ls = "--")
    x0 = x1

pylab.plot( x , y  )
 
pylab.show()
```

```python
n = int ( input("> ") )

t = 2*n-1 if n%2 else 2*n+1
x = [ "" for i in range ( t ) ]
y = [ "" for i in range ( t ) ]

r , s = [ 0 , 1 , 0 , -1 ] , [ -1 , 0 , 1 , 0 ]
a , b , c = 0 , 0 , 0

for i in range ( n**2 + n ) :

    if a == b == 0 or ( a + b == n - 1 and b < a ) or a - b == -1 or ( a + b == n and b > a )  :
    
        x[c] = a
        y[c] = - b
        c = c + 1
    
    a = a + r[ c%4 ] 
    b = b + s[ c%4 ]

color = [ "c" , "purple" , "g" , "r" ]

r , s = [ 0 , -1 , 0 , 1 ] , [ 0 , -1 , -2 , -1 ]

for i in range (4) :

    d = (i+1)/2
    q = array( [ [ cos( d*pi )  , -sin( d*pi ) ] , [ sin( d*pi ), cos( d*pi )]  ] ) #選轉公式
    w = [ x , y ]
    e = dot( q , w ) + array( [ [ r[i] ] , [ s[i] ] ] )

    pylab.plot( e[0] , e[1] , color = color[i] )

pylab.plot( 0 , -1 ,  marker = "o" ,  color = "black" )#球
pylab.axis( "off" ) #顯示區域(前面加"off"關掉做標軸)
pylab.show()

```

```python
a , b  = eval( input("時,分> ") )

n1 , n2 = ( -a*5 - b*5/60 - 15 )/30*pi , ( -b - 15 )/30*pi 
s = [ 4 , 4.1 , 0.15 ]
ts = linspace( 0 , 10*pi , 1000 )

for i in  s : #外面兩圓 加 裡面綠色的圓
    x , y = cos( ts )*i , sin( ts )*i
    pylab.plot( x , y , color = "black" if i != 0.15 else "green"  )

for i in range( 60 ) : 
    a = i/30*pi
    pylab.arrow( cos(a)*4  , sin(a)*4 , -cos(a)*0.5 , -sin(a)*0.5 , color = "blue" )

for i in range( 12 ) :
    a = i/6*pi
    pylab.arrow( cos(a)*4  , sin(a)*4 , -cos(a) , -sin(a) , color = "red" )

    if ( 15 - i )%3 == 0 :
        pylab.text( cos(a)*2.5 , sin(a)*2.5 , ( 12 - i )%12 + 3   , color = "red" )

pylab.arrow( 0  , 0 , -cos( n1 )*1.5 , -sin( n1 )*1.5, color = "black" ) #短針
pylab.arrow( 0  , 0 , -cos( n2 )*2.5 , -sin( n2 )*2.5, color = "black" ) #長針

pylab.axis( "off" ) #顯示區域(前面加"off"關掉做標軸)
pylab.show()
```
