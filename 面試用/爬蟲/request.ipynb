```python
import urllib.request as request
import json
```

```python
#中央大學
url = "https://www.ncu.edu.tw/"
with request.urlopen(url) as indata : 
    for x in indata : 
        print( x.decode('utf-8') , end="" )
```

```python
#臺北市電動機車充電地址
url = "https://data.taipei/api/v1/dataset/c2b666e0-f848-4609-90cb-5e416435b93a?scope=resourceAquire"
with request.urlopen(url) as indata :  
    data = json.load( indata )
```

```python
for name in data['result'] : 
    print( name , end=" " )
```

```python
print( data['result']['sort'] )
dic = { x['行政區']  : [] for x in data['result']['results'] } 
for x in data['result']['results'] : 
    dic[ x['行政區'] ] += [x['地址']]
    
for x in dic : 
    print( x )
    for y in dic[x] : 
        print( y , end=" " )
    print()
```
